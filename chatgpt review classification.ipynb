```python
# Sentiment analysis
def analyze_sentiment(reviews):
    batch_size = 5
    review_batches = [reviews[i:i + batch_size] for i in range(0, len(reviews), batch_size)]

    # Perform sentiment analysis for each batch
    for batch in review_batches:
        # Construct prompt for each review in the batch
        prompts = [f"Sentiment analysis: {review}" for review in batch]

        # Make the API request
        response = openai.Completion.create(
            engine="text-davinci-003",  
            prompt=prompts,
            max_tokens=50,  
            n=len(batch),
            stop=None,
            temperature=0.7,  
        )

       
        outputs = [choice['text'].strip() for choice in response['choices']]
        
        for i, review in enumerate(batch):
            print(f"Review: {review}")
            print(f"Sentiment: {outputs[i]}")
            print("=" * 30)
```
